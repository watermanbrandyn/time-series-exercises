2. Make a scatterplot of displacement vs horsepower. Add the keyword argument hue='weight'. What does this change? Add the keyword argument palette='Greens'. What does this change?

````python
sns.scatterplot(x='displacement', y='horsepower', data=df)
````

````python
sns.scatterplot(x='displacement', y='horsepower', data=df, hue='weight')
````

````python
sns.scatterplot(x='displacement', y='horsepower', data=df, hue='weight', palette='Greens')
````

````python
# Hue applies another column to categorize data with (in this case a continuous column), palette changes the 
# color scheme for this
````

3. Make a scatterplot of displacement vs mpg. Add the keyword argument hue='cylinders'. What does this change? Add the line of code below in a seperate cell, run it, and then create your scatterplot again. What does this change?

````python
sns.scatterplot(x='mpg', y='displacement', data=df)
````

````python
df.cylinders = df.cylinders.astype('category')
````

````python
sns.scatterplot(data=df, y='mpg', x='displacement', hue='cylinders')
````

````python
sns.relplot(data=df, y='mpg', x='displacement', col='origin', row='cylinders')
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13,7))
sns.scatterplot(data=df, y='mpg', x='displacement', ax=ax1)
sns.scatterplot(data=df, y='mpg', x='weight', ax=ax2)

ax1.set(title='Diplacement v MPG')
ax2.set(title='Weight v MPG')
fig.suptitle('Cars MPG Viz')
````
